<html>
    <body>
      <p style = "font-family:courier new,serif;font-size:18px;font-style:bold;">
         <b>Logistic Regression</b>
          <br>Logistic Regression is a Classifiaction model having linear function of Indepent Variables</br>
          <br>__init__ is defined to intilaize the variables </br>
          <br> Logistic Regression uses Sigmoid function as it scales down the Output form 0 to 1 so that an threshold can be set such that above which is desired Output and below the threshold is considered underised output hence 'sigmoid' function is defined</br>
          <br>In 'model' W is Weights which is the coeficients of those independent variables which is initialized to Zero</br>
          <br>An for loop is put to iterate for all the train samples of Wine </br>
          <br>H is the Linear function while Y_predicted is the Sigmoid function of H</br>
          <br>cost function is average of error that occurs during prediction</br>
          <br>dW is derivative of cost fucntion with respect to W while dB is derivative of cost fucntion with respect to B</br>
          <br>Next step updates the Values of Weights(W) and B</br>
          <br>In 'pdt' X_test data is passed to predict the Outcome and the Output is sorted as mentioned above according to the threshold which 0.45 </br>
      </p>
   </body>
</html>

```python
class Logistic_Regression :
    
    def __init__ (self,learning_rate ,iterations ) :
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.W = 0
        self.B = 0
    
    def sigmoid(self,x) :         #Defines Sigmoid Fucntion
        return 1/(1 + np.exp(-x))
    
    def model(self,X,Y):    #Gradient Descent
        
        self.W = np.zeros(X.shape[1])
        self.B = 0
        
        for i in range(self.iterations):
            
            H = np.dot(X,self.W) + self.B
            Y_predicted = self.sigmoid(H)
            
            dW = (1/X.shape[0])*np.dot(X.T,(Y_predicted - Y))
            dB = (1/X.shape[0])*np.sum(Y_predicted - Y)
            
            self.W = self.W - dW*self.learning_rate
            self.B = self.B - dB*self.learning_rate
            
    def pdt(self,X):    #Pred
        
        H = np.dot(X,self.W) + self.B
        Y_predicted = self.sigmoid(H)
        
        Y_predicted_f = [1 if i > 0.45 else 0 for i in Y_predicted]
            
        return np.array(Y_predicted_f)
```

<html>
    <body>
      <p style = "font-family:courier new,serif;font-size:18px;font-style:bold;">
         'accuracy' is defined such that it returns _accuracy which is the ratio of Number of Correctly predicted sample vs Total number of samples in the test dataset
      </p>
   </body>
</html>

```python
def accuracy(Y_test,Y_predicted_f):
        
    n = sum (Y_test == Y_predicted_f)
    _accuracy = n/len(Y_test)
    
    return _accuracy

A = Logistic_Regression(learning_rate = 1 ,iterations = 10000)
A.model(X_train,Y_train)
B = A.pdt(X_test)

print("Logistic Classification Accuracy is : ",accuracy(Y_test,B)*100," %")  
```
